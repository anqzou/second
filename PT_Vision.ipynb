# 图片正则化

```python
def normalization (image,width,height):
    image_copy = np.copy(image)
    image_normalize = cv.resize(image_copy,(width,height))
    return image_normalize
```

# 热编码

```python
def file(label):
    classify = [0,1]
    if label =="clean_list":
        classify = [1,0]
    return classify
```
